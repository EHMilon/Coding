# VSCode Keyboard Shortcuts

- Add Line Comment: `Ctrl+/`
- Block Comment: `Alt+/`

- Copy Line Down: `Ctrl+Down Arrow`
- Copy Line Up: `Ctrl+Up Arrow`

- Move Line Down: `Alt+Down Arrow`
- Move Line Up: `Alt+Up Arrow`
- Move Selected Text Left: `Alt+Left Arrow`
- Move Selected Text Right: `Alt+Right Arrow`

- Delete Line: `Ctrl+Backspace`
- Delete Word: `Shift+Backspace`
- Remove Brackets: `Alt+Backspace`
- Delete All to the Left: `Ctrl+Left Arrow+Backspace`
- Delete All to the Right: `Ctrl+Right Arrow+Backspace`

- Indent Line: `Ctrl+Tab`
- Outdent Line: `Shift+Tab`

- Insert Line Below: `Shift+Enter`
- Insert Line Above: `Ctrl+Shift+Enter`

- View: Move Editor into Preview Group: `Ctrl+Alt+Right Arrow`
- View: Move Editor into Next Group: `Ctrl+Alt+Left Arrow`

- Spell: Add Words to Dictionary: `Ctrl+\``
- Spell: Spelling Suggestion: `Ctrl+X`

- Cursor Undo: `Ctrl+U`
- Cursor Redo: `Ctrl+Shift+U`
- Cursor to Top of Page: `Ctrl+Home`
- Cursor to Bottom of Page: `Ctrl+End`

- Run Code: `Shift+Space`
- View: Close Editor: `Ctrl+W`
- Open New External Terminal: `Ctrl+T`
- Trigger Suggest: `Ctrl+I` or `Ctrl+Space`
- Terminal Clear: `Delete`
- Copy Path of Active File: `Shift+Alt+C`
- Expand Line Selection: `Ctrl+L`
- File: Compare Active File with Clipboard: `Alt+V`

- Select All Occurrences of Find Match: `Alt+A`
- Select from Anchor to Cursor: Press and hold the `Shift` key
- Add Selection to Next Find Match: `Ctrl+D`
- Add Selection to Preview Find Match: `Ctrl+Right Arrow`
- Move Last Selection to Next Find Match: `Ctrl+Left Arrow`

- File: New File (Explorer): `Alt+N`
- File: New Folder (Explorer): `Ctrl+Alt+N`
- File: Open File: `Ctrl+O`
- File: Open Folder: `Alt+O`
- File: Open Recent: `Ctrl+Alt+0`
- Rename Symbol: `F2`
- Delete File: `Delete`

- Create: New Jupyter Notebook: `Ctrl+Shift+J`
- Developer: Inspect Editor Tokens and Scopes: `Ctrl+F2`
- Simple Browser: `Alt+B`

- Terminal: Move Terminal into Editor Area: `Ctrl+E, T`
- Terminal: Move Terminal into Panel Area: `Alt+E, T`
- Terminal: Toggle Size to Content Width: `Alt+Z`

- Transform to Lowercase: `Ctrl+Shift+L`
- Transform to Uppercase: `Ctrl+Shift+U`
- Transform to Title Case: `Ctrl+Shift+T`

- Create Folding Range from Selection: `Ctrl+[`
- Unfold: `Ctrl+]`
- Remove Manual Folding Ranges: `Ctrl+]`

- Reload Window: `Ctrl+R`

- Markdown: Open Preview: `Ctrl+M`
- Word Wrap: `Alt+Z`
- Replace: `Ctrl+F`
- Compare File: `Alt+C`
